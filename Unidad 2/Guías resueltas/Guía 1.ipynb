### Script 1

Desarrolla un script que pida un número al usuario e identifique si el valor ingresado es positivo, negativo o cero.

```python
# Identificardor de números positivos y negativos

# Validador de entrada
while True:
    try:
        numero = float(input("Ingrese un número: "))
        break
    except ValueError:
        print("Entrada no válida. Por favor, ingrese un número.")
        
# Identificación de números positivos, negativos o cero
if numero > 0:
    print(f"El número {numero} es positivo.")
elif numero < 0:
    print(f"El número {numero} es negativo.")
else:
    print("El número es cero.")
```

Desarrolla un script que pida dos números al usuario y determine cuál de ellos es mayor, o bien indique si ambos son iguales.

```python
# Ideentificador de núemeros mayores y menores.

# Validador de entrada
while True:
    try:
        numero1 = float(input("Ingrese el primer número: "))
        numero2 = float(input("Ingrese el segundo número: "))
        break
    except ValueError:
        print("Entrada no válida. Por favor, ingrese números.")

# Identificación del número mayor y menor
if numero1 > numero2:
    print(f"El número {numero1} es mayor que {numero2}.")
elif numero1 < numero2:
    print(f"El número {numero1} es menor que {numero2}.")
else:
    print(f"Los números {numero1} y {numero2} son iguales.")
```

### Script 4.
Desarrolla un script que pida dos números enteros al usuario e indique si el primer número es múltiplo del segundo, utilizando el operador módulo (%).

```python
# Identificador de múltiplos

# Validador de entrada
while True:
    try:
        numero = float(input("Ingrese un número: "))
        multiplo = float(input("Ingrese un número para ver si es múltiplo del número anterior: "))
        break
    except ValueError:
        print("Entrada no válida. Por favor, ingrese números.")

# Identificación de múltiplos
if numero % multiplo == 0:
    print(f"El número {numero} es múltiplo de {multiplo}.")
else:
    print(f"El número {numero} no es múltiplo de {multiplo}.")
```

### Script 5

Desarrolla un script que pida la edad de una persona e indique si corresponde a menor de edad, mayor de edad, adulto mayor o cuarta edad en Chile. Para ello, el estudiante debe investigar previamente los rangos etarios y aplicarlos correctamente en el script.

```python
'''
Código para ver si una persona es mayor de edad, menor de edad o adulto mayor.

En chile, la clasificación de edades es la siguiente:
- Menor de edad: 0 a 17 años
- Mayor de edad: 18 a 60 años
- Adulto mayor: 60 años o más
'''

print("--- Clasificación de edades ---")

# Validador de entrada
while True:
    try:
        edad = int(input("Ingrese su edad: "))
        break
    except ValueError:
        print("Entrada no válida. Por favor, ingrese un número entero.")

# Validador de edades validas
while True:
    if edad < 0:
        print("La edad no puede ser negativa. Por favor, ingrese una edad válida.")
        try:
            edad = int(input("Ingrese su edad: "))
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número entero.")
    else:
        break

# Clasificación de edades
if edad < 18:
    print("Eres menor de edad.")
elif 18 <= edad < 60:
    print("Eres mayor de edad.")
else:
    print("Eres adulto mayor.")
```

### Script 7

Desarrolla un script que pida una nota final en escala de 1.0 a 7.0 e indique si el estudiante está aprobado o reprobado, considerando que en Chile se aprueba con nota 4.0 o superior.

```python
'''

Clasificador de notas.

La clasificación es la siguiente:
- Reprobado: Menos de 4.0
- Aprobado: 4.0 a 7.0

'''

print("--- Clasificación de notas ---")

# Validador de entrada
while True:
    try:
        nota = float(input("Ingrese la nota del estudiante (1.0 a 7.0): "))
        break
    except ValueError:
        print("Entrada no válida. Por favor, ingrese un número.")
        
# Validador de notas validas
while True:
    if nota < 1.0 or nota > 7.0:
        print("La nota debe estar entre 1.0 y 7.0. Por favor, ingrese una nota válida.")
        try:
            nota = float(input("Ingrese la nota del estudiante (1.0 a 7.0): "))
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número.")
    else:
        break

# Clasificación de notas
if nota < 4.0:
    print(f"El estudiante con nota {nota},  ha reprobado.")
else:
    print(f"El estudiante con nota {nota}, ha aprobado.")
```

### Script 8.

Desarrolla un script que pida el nombre de un usuario y su año de nacimiento, y luego muestre un mensaje indicando el nombre de la persona y la edad que tiene actualmente

```python
'''

Código que pide el nombre de una persona y su año de nacimiento. Luego muestra el nombre y la edad actual de la persona.

La forma de saber la edad actual de la persona es restando el año de nacimiento al año actual. 
Para obtener el año actual, se puede usar la función datetime del módulo datetime.
    - O poner el año actual manualmente, pero esto no es recomendable ya que el código no sería reutilizable en el futuro.
    
Ahora, para realizar el cálculo de edad más preciso, se podría pedir también el mes y el día de nacimiento, 
y compararlo con la fecha actual para determinar si la persona ya ha cumplido años en el año actual o no. 
Sin embargo, para simplificar el código, solo se pedirá el año de nacimiento y se asumirá que la persona 
ya ha cumplido años en el año actual.

'''

import datetime

año_actual = datetime.datetime.now().year

print("--- Cálculo de edad ---")

# Validador de entrada
while True:
    try:
        nombre = input("Ingrese su nombre: ")
        año_nacimiento = int(input("Ingrese su año de nacimiento: "))
        break
    except ValueError:
        print("Entrada no válida. Por favor, ingrese un número entero para el año de nacimiento.")
        
# Validador de año de nacimiento válido
while True:
    if año_nacimiento < 0 or año_nacimiento > año_actual:
        print(f"El año de nacimiento debe estar entre 0 y {año_actual}. Por favor, ingrese un año válido.")
        try:
            año_nacimiento = int(input("Ingrese su año de nacimiento: "))
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número entero para el año de nacimiento.")
    else:
        break

# Cálculo de edad
edad = año_actual - año_nacimiento

# Mostrar resultados
print(f"Nombre: {nombre}")
print(f"Edad: {edad} años")
```

### Script 9

### Script 10.

Desarrolla un script que considere un producto con precio base de $80.000. Luego, debe pedir al usuario la forma de pago: efectivo, débito o crédito.

Aplica las siguientes reglas:

- Si paga con efectivo, se realiza un descuento de $5.000.
- Si paga con crédito, se aplica un recargo de $3.000.
- Si paga con débito, el precio no cambia.

El script debe mostrar el precio final a pagar.

```python
'''

Código que pide la forma de pago y aplica un descuento o recargo según la forma de pago.

Sabemos el precio base es $80.000

Según el método de pago, se aplican los siguientes descuentos o recargos:
    - Efectivo: $5.000 de descuento
    - Tarjeta de crédito: $3.000 de recargo
    - Débito: no se aplica ningún descuento ni recargo
    
'''

print("--- Cálculo de forma de pago ---")
precio_base = 80000

# Validador de entrada
while True:
    try:
        forma_pago = input("Ingrese la forma de pago (efectivo, tarjeta de crédito, débito): ").lower()
        break
    except ValueError:
        print("Entrada no válida. Por favor, ingrese una forma de pago válida.")
        
# Validador de formas de pago válidas
while True:
    if forma_pago not in ["efectivo", "tarjeta de crédito", "débito"]:
        print("Forma de pago no válida. Por favor, ingrese una forma de pago válida.")
        try:
            forma_pago = input("Ingrese la forma de pago (efectivo, tarjeta de crédito, débito): ").lower()
        except ValueError:
            print("Entrada no válida. Por favor, ingrese una forma de pago válida.")
    else:
        break

# Cálculo del precio final
if forma_pago == "efectivo":
    precio_final = precio_base - 5000
    print("El precio base es de $80.000, pero al pagar en efectivo se aplica un descuento de $5.000.")
    print(f"Usted pagará ${precio_final} en efectivo.")
elif forma_pago == "tarjeta de crédito":
    precio_final = precio_base + 3000
    print("El precio base es de $80.000, pero al pagar con tarjeta de crédito se aplica un recargo de $3.000.")
    print(f"Usted pagará ${precio_final} con tarjeta de crédito.")
else:
    precio_final = precio_base
    print("El precio base es de $80.000, pero al pagar con débito no se aplica ningún descuento ni recargo.")
    print(f"Usted pagará ${precio_final} con débito.")
```

### Script 11.

Desarrolla un script para una compañía de arriendo de autos que cobra una tarifa fija de $50.000 por día.

Debes pedir al usuario la cantidad de días de arriendo y calcular el valor final según estas reglas:
- Si arrienda más de 7 días, obtiene un 25% de descuento sobre el total.
- Si arrienda más de 3 días y hasta 7 días, obtiene un 10% de descuento sobre el total.
- Si arrienda 3 días o menos, no recibe descuento.

El script debe mostrar el total a pagar

```python
'''

Código para una compañía de autos.

Sabemos que el precio base es $50.000

    - Más de 7 días = 50.000 * 0.75
    - Entre 3 y 7 días = 50.000 * 0.90
    - 3 o menos de 3 días = 50.000
    
Hay que pedir cantidad de días que se va a arrendar el auto y mostrar el precio final a pagar.

'''

print("--- Cálculo de arriendo de auto ---")
precio_base = 50000

# Validador de entrada
while True:
    try:
        dias_arriendo = int(input("Ingrese la cantidad de días que va a arrendar el auto: "))
        break
    except ValueError:
        print("Entrada no válida. Por favor, ingrese un número entero para los días de arriendo.")

# Validador de días de arriendo válidos
while True:
    if dias_arriendo < 0:
        print("La cantidad de días no puede ser negativa. Por favor, ingrese un número válido.")
        try:
            dias_arriendo = int(input("Ingrese la cantidad de días que va a arrendar el auto: "))
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número entero para los días de arriendo.")
    else:
        break

# Cálculo del precio final
if dias_arriendo <=3:
    precio_final = precio_base
    print(f"Al arrendar el auto por {dias_arriendo} días, no se aplica ningún descuento.")
    print(f"Usted pagará ${precio_final} por el arriendo del auto.")
elif 3 < dias_arriendo <= 7:
    precio_final = int(precio_base * 0.90)
    print(f"Al arrendar el auto por {dias_arriendo} días, se aplica un descuento del 10%.")
    print(f"Usted pagará ${precio_final} por el arriendo del auto.")
else:
    precio_final = int(precio_base * 0.75)
    print(f"Al arrendar el auto por {dias_arriendo} días, se aplica un descuento del 25%.")
    print(f"Usted pagará ${precio_final} por el arriendo del auto.")
```

### Script 12

Desarrolla un script para una tienda que aplica descuentos según el monto de compra. Debes pedir al usuario el monto total de la compra y calcular el valor final a pagar considerando las siguientes reglas:
- Si la compra es mayor a $40.000 y hasta $100.000, se aplica un 10% de descuento.
- Si la compra es superior a $100.000, se aplica un 20% de descuento.
- Si la compra es de $40.000 o menos, no se aplica descuento.

El script debe mostrar el monto original, el descuento aplicado y el total a pagar

```python
'''

Cálculo de descuento por cantidad de productos comprados.

Se pide el total de la compra, según el total se aplican los siguientes descuentos:
    - Más de $100.000 = 20% de descuento
    - Entre $40.000 y $100.000 = 10% de descuento
    - Menos de $40.000 = no se aplica ningún descuento
    
Se pide el total de compra y se muestra el precio final a pagar.

'''

print("--- Cálculo de descuento por compra ---")

# Validador de entrada
while True:
    try:
        total_compra = float(input("Ingrese el total de la compra: "))
        break
    except ValueError:
        print("Entrada no válida. Por favor, ingrese un número para el total de la compra.")
        
# Validador de total de compra válido
while True:
    if total_compra < 0:
        print("El total de la compra no puede ser negativo o decimal. Por favor, ingrese un número válido.")
        try:
            total_compra = float(input("Ingrese el total de la compra: "))
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número para el total de la compra.")
    else:
        break
    
# Cálculo del precio final
if total_compra > 100000:
    precio_final = float(total_compra * 0.80)
    print(f"Al tener un total de compra de ${total_compra}, se aplica un descuento del 20%.")
    print(f"Usted pagará ${precio_final} por su compra.")
elif 40000 <= total_compra <= 100000:
    precio_final = float(total_compra * 0.90)
    print(f"Al tener un total de compra de ${total_compra}, se aplica un descuento del 10%.")
    print(f"Usted pagará ${precio_final} por su compra.")
else:
    precio_final = total_compra
    print(f"Al tener un total de compra de ${total_compra}, no se aplica ningún descuento.")
    print(f"Usted pagará ${precio_final} por su compra.")
```

### Script 13

Desarrolla un script que pida tres números al usuario y determine cuál de ellos es el mayor. Si existen números iguales, el programa también debe indicarlo

```python
# Código que pide tres números, identifica el mayor y el menor o si hay alguno igual, y muestra el resultado.

print("--- Identificación del número mayor y menor ---")

# Validador de entrada
while True:
    try:
        num1 = float(input("Ingrese el primer número: "))
        break
    except ValueError:
        print("Entrada no válida. Por favor, ingrese un número.")

while True:
    try:
        num2 = float(input("Ingrese el segundo número: "))
        break
    except ValueError:
        print("Entrada no válida. Por favor, ingrese un número.")

while True:
    try:
        num3 = float(input("Ingrese el tercer número: "))
        break
    except ValueError:
        print("Entrada no válida. Por favor, ingrese un número.")

# Identificación del mayor y el menor
mayor = max(num1, num2, num3)
menor = min(num1, num2, num3)

# Verificación de números iguales
if num1 == num2 == num3:
    print("Los tres números son iguales.")
elif num1 == num2 or num1 == num3 or num2 == num3:
    print("Hay números iguales, pero no todos son iguales.")
# Notar que las instrucciones no especifican qué hacer en caso de números iguales, 
# por lo que se asume que se muestra el mayor y el menor aunque haya números iguales.

# Mostrar resultado
print(f"El número mayor es: {mayor}")
print(f"El número menor es: {menor}")
```
