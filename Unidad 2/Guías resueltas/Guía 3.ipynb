### Ejercicio 1: Análisis de pesos en salmonicultura

En el sur de Chile, la industria del salmón requiere monitorear constantemente el peso de los peces para tomar
decisiones de alimentación y crecimiento.

Desarrolla un script en Python que permita analizar una muestra de salmones.

El programa debe:
- Generar una lista con 10 pesos de salmones (puedes pedirlos al usuario o generarlos automáticamente usando random en un rango entre 2.5 y 6.0 kg).
- Mostrar:
    + El peso promedio (promedio = suma de pesos / cantidad de salmones)
    + El peso más bajo
    + El peso más alto
    + Cuántos salmones están sobre el promedio

Consideraciones:
- No usar funciones como max(), min() o sum()
- Usar ciclos (for o while)
- Separar el cálculo del promedio del análisis posterior


```python
'''

Código que analiza muestra de salmones.

El peso de los salmones es de 2.5 kg a 6.0 kg.

Se pide que se ingrese el peso de 10 salmones y se guardan en una lista.

Según el peso, se debe mostrar:
    - Peso promedio.
    - El peso más bajo.
    - El peso más alto.
    - Cuantos salmones hay sobre el peso promedio.
    
Reglas:
    - No usar funciones como max(), min() o sum().
    - Usar ciclos.
    - Separar el cálculo del promedio de los demás cálculos.

'''

print("\n--- Análisis de Salmones ---")

#Función para validar el peso del salmón
def validar_peso(peso):
    if 2.5 <= peso <= 6.0:
        return True
    else:
        return False

#Lista para almacenar los pesos de los salmones
pesos_salmones = []

#Ingreso de pesos de los salmones
for i in range(10):
    while True:
        try:
            peso = float(input(f"Ingrese el peso del salmón {i+1} (2.5 kg a 6.0 kg): "))
            if validar_peso(peso):
                pesos_salmones.append(peso)
                break
            else:
                print("Peso inválido. Por favor, ingrese un peso entre 2.5 kg y 6.0 kg.")
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número.")
            
#Cálculo del peso promedio
suma_pesos = 0
for peso in pesos_salmones:
    suma_pesos += peso
promedio = suma_pesos / len(pesos_salmones)

#Cálculo del peso más bajo y más alto
peso_mas_bajo = pesos_salmones[0]
peso_mas_alto = pesos_salmones[0]
for peso in pesos_salmones:
    if peso < peso_mas_bajo:
        peso_mas_bajo = peso
    if peso > peso_mas_alto:
        peso_mas_alto = peso
        
#Cálculo de cuantos salmones hay sobre el peso promedio
salmones_sobre_promedio = 0
for peso in pesos_salmones:
    if peso > promedio:
        salmones_sobre_promedio += 1

#Resultados
print("\n--- Resultados ---")
print(f"Peso promedio: {promedio:.2f} kg")
print(f"Peso más bajo: {peso_mas_bajo:.2f} kg")
print(f"Peso más alto: {peso_mas_alto:.2f} kg")
print(f"Cantidad de salmones sobre el peso promedio: {salmones_sobre_promedio}")
```

### Ejercicio 2: Control de temperatura en invernadero agrícola

En la agricultura chilena, especialmente en zonas donde se producen hortalizas bajo invernadero, es importante controlar la temperatura para evitar pérdidas en los cultivos.

Desarrolla un script en Python que permita analizar las temperaturas registradas durante una jornada dentro de un invernadero.

El programa debe:
- Crear una tupla con 12 temperaturas registradas durante el día. Puedes ingresarlas manualmente o generarlas usando random en un rango entre 8 y 35 grados.
- Recorrer la tupla y mostrar:
    + La temperatura más baja.
    + La temperatura más alta.
    + Cuántas mediciones estuvieron bajo los 12 grados.
    + Cuántas mediciones estuvieron sobre los 30 grados.

Consideraciones:
- Usa una tupla porque las mediciones ya fueron registradas y no deberían modificarse.
- Se recomienda calcular manualmente el promedio, mínimo y máximo para practicar recorridos.
- El programa debe mostrar un pequeño mensaje final, por ejemplo: “Jornada estable”, “Riesgo por frío” o “Riesgo por exceso de calor”.


```python
''' 

Cpodigo que analiza las temperaturas de un cultivo de invernadero.

Se debe generar un tupla con el registro de 12 temperaturas, entre 8 y 35 grados Celsius. 

Según la tupla se debe mostrar: 
    - Temperatura promedio.
    - La temperatura más baja.
    - La temperatura más alta.
    - Cuantas mediciones estuvieron los 12 grados.
    - Cuantas mediciones estuvieros sobre los 30 grados.
    
Se debe mostrar un mensaje final según el resultado del análisis:
    - Si la temperatura promedio es menor a 15 grados, mostrar "Temperatura baja"
    - Si la temperatura promedio es entre 15 y 35 grados, mostrar "Temperatura óptima"
    - Si la temperatura promedio es mayor a 35 grados, mostrar "Temperatura alta"

'''

print("\n--- Análisis de Temperaturas en Invernadero ---")

#Función para validar la temperatura
def validar_temperatura(temperatura):
    if 8 <= temperatura <= 35:
        return True
    else:
        return False
    
#Tupla para almacenar las temperaturas
temperaturas = ()

#Ingreso de temperaturas
for i in range(12):
    while True:
        try:
            temperatura = float(input(f"Ingrese la temperatura {i+1} (8°C a 35°C): "))
            if validar_temperatura(temperatura):
                temperaturas += (temperatura,)
                break
            else:
                print("Temperatura inválida. Por favor, ingrese una temperatura entre 8°C y 35°C.")
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número.")

#Cálculo de la temperatura promedio
suma_temperaturas = 0
for temperatura in temperaturas:
    suma_temperaturas += temperatura

promedio_temperatura = suma_temperaturas / len(temperaturas)

#Cálculo de la temperatura más baja y más alta
temperatura_mas_baja = temperaturas[0]
temperatura_mas_alta = temperaturas[0]
for temperatura in temperaturas:
    if temperatura < temperatura_mas_baja:
        temperatura_mas_baja = temperatura
    if temperatura > temperatura_mas_alta:
        temperatura_mas_alta = temperatura

#Cálculo de las mediciones bajo los 12 grados
mediciones_bajo_12 = 0
for temperatura in temperaturas:
    if temperatura < 12:
        mediciones_bajo_12 += 1

#Cálculo de las mediciones sobre 30 grados
mediciones_sobre_30 = 0
for temperatura in temperaturas:
    if temperatura > 30:
        mediciones_sobre_30 += 1

#Mostrar resultados
print(f"\n--- Resultados del Análisis ---")
print(f"Temperatura promedio: {promedio_temperatura:.2f}°C")
print(f"Temperatura más baja: {temperatura_mas_baja}°C")
print(f"Temperatura más alta: {temperatura_mas_alta}°C")
print(f"Mediciones bajo 12°C: {mediciones_bajo_12}")
print(f"Mediciones sobre 30°C: {mediciones_sobre_30}")

#Mensaje final según el resultado del análisis
if promedio_temperatura < 15:
    print("Temperatura baja, hay riesgo por frío en el invernadero")
elif 15 <= promedio_temperatura <= 35:
    print("Temperatura óptima, el invernadero está en condiciones ideales")
else:
    print("Temperatura alta, hay riesgo por exceso de calor en el invernadero")
```

### Ejercicio 4: Registro de ventas en feria agrícola

En una feria agrícola local, distintos productores venden frutas y verduras durante la mañana. 

El encargado necesita registrar las ventas realizadas para conocer el comportamiento del día.

Desarrolla un script en Python que permita registrar ventas hasta que el usuario decida terminar.

El programa debe:
+ Usar un diccionario donde la clave sea el nombre del producto y el valor sea la cantidad total vendida.
+ Solicitar repetidamente al usuario:
    + nombre del producto vendido
    + cantidad vendida
- Si el producto ya existe en el diccionario, se debe sumar la nueva cantidad a la cantidad anterior.
- El ingreso de ventas debe continuar hasta que el usuario escriba "fin".
- Al finalizar, mostrar:
    + todos los productos vendidos con sus cantidades
    + el producto más vendido
    + la cantidad total de unidades vendidas entre todos los productos

Consideraciones:
- Este ejercicio debe usar un ciclo de duración desconocida, por ejemplo while.
- Dentro del while, se puede recorrer el diccionario con for para obtener los resultados finales.
- Se recomienda calcular manualmente el producto más vendido para practicar comparación y recorrido

```python
'''

Registro de ventas en feria agrícola.

El código debe generar un diccionario donde la clave sea el nombre del producto y el valor sea la 
cantidad total vendida.

Se debe ejecutar un ciclo indeterminado que solicite:
    - Nombre de producto vendido.
    - Cantidad vendida.
        - Si el producto ya existe en el diccionario, se debe sumar la cantidad vendida a la cantidad existente.

El ciclo debe finalizar cuando el usuario ingrese "fin" como nombre de producto.

Se debe mostrar:
    - todos los productos vendidos con sus cantidades
    - el producto más vendido
    - la cantidad total de unidades vendidas entre todos los productos
'''

print("\n--- Registro de Ventas en Feria Agrícola ---")

#Función para validar la cantidad vendida
def validar_cantidad(cantidad):
    if cantidad >= 0:
        return True
    else:
        return False

#Diccionario para almacenar las ventas
ventas = {}

#Ciclo indeterminado para ingresar las ventas
while True:
    nombre_producto = input("Ingrese el nombre del producto vendido (o 'fin' para terminar): ")
    if nombre_producto.lower() == "fin":
        break
    while True:
        try:
            cantidad_vendida = int(input(f"Ingrese la cantidad vendida de {nombre_producto}: "))
            if validar_cantidad(cantidad_vendida):
                if nombre_producto in ventas:
                    ventas[nombre_producto] += cantidad_vendida
                else:
                    ventas[nombre_producto] = cantidad_vendida
                break
            else:
                print("Cantidad inválida. Por favor, ingrese un número positivo.")
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número entero.")

# Mostrar los resultados
print("\nResultados:")
for producto, cantidad in ventas.items():
    print(f"Se vendieron {cantidad} unidades de {producto}")

if ventas:
    producto_mas_vendido = max(ventas, key=ventas.get)
    print(f"\nProducto más vendido: {producto_mas_vendido} con {ventas[producto_mas_vendido]} unidades vendidas.")
    
    '''
    Nota: La función key=ventas.get en max() permite encontrar la clave del diccionario que tiene el valor máximo, 
    es decir, el producto más vendido.
    
    Sin la función key=ventas.get, max(ventas) simplemente devolvería la clave con el valor máximo en orden alfabético, 
    lo cual no es correcto para este caso.
    '''

cantidad_total = sum(ventas.values())
print(f"Cantidad total de unidades vendidas: {cantidad_total}")
```

### Ejercicio 5: Monitoreo de niveles de oxígeno en centros de cultivo

En los centros de cultivo de salmones, el nivel de oxígeno disuelto en el agua es crítico para la supervivencia de los peces. Durante el día, un sensor registra mediciones periódicas.

Desarrolla un script en Python que permita analizar estas mediciones.

El programa debe:
- Generar una lista con valores de oxígeno disuelto (puedes usar random en un rango entre 4.0 y 10.0 mg/L o ingresarlos manualmente).
- La cantidad de mediciones no es fija: el usuario debe indicar cuántas mediciones se registrarán.
- Mostrar:
    + El promedio de oxígeno (promedio = suma de mediciones / cantidad de mediciones)
    + Cuántas mediciones están bajo 5.0 mg/L (nivel crítico)
    + Cuántas están entre 5.0 y 7.0 mg/L (nivel aceptable)
    + Cuántas están sobre 7.0 mg/L (nivel óptimo)

Consideraciones:
- Debes usar un ciclo para generar o ingresar las mediciones (por ejemplo, un while o for según corresponda).
- Luego, debes recorrer la lista para realizar el análisis.
- Piensa cómo clasificar cada medición en su categoría correspondiente.


```python
'''

Monitoreo de nivel de oxígeno en un cultivo de salmones.

El código debe generar una lista con los valores de oxígene disuelto.
    - Los niveles de oxígeno deben estar entre 4.0 y 10.0 mg/L
Se debe pedir la cantidad de mediciones a ingresar y luego solicitar cada medición, 
validando que esté dentro del rango permitido.

Se debe mostrar:
    - El promedio de oxígeno (promedio = suma de mediciones / cantidad de mediciones)
    - Cuántas mediciones están bajo 5.0 mg/L (nivel crítico)
    - Cuántas están entre 5.0 y 7.0 mg/L (nivel aceptable)
    - Cuántas están sobre 7.0 mg/L (nivel óptimo)

'''

print("\n--- Monitoreo de Nivel de Oxígeno en Cultivo de Salmones ---")

#Función para validar el nivel de oxígeno
def validar_oxigeno(oxigeno):
    if 4.0 <= oxigeno <= 10.0:
        return True
    else:
        return False

#Lista para almacenar los niveles de oxígeno
niveles_oxigeno = []

#Ingreso de la cantidad de mediciones y validación
while True:
    try:
        cantidad_mediciones = int(input("Ingrese la cantidad de mediciones a ingresar: "))
        if cantidad_mediciones > 0:
            break
        else:
            print("La cantidad de mediciones debe ser un número positivo.")
    except ValueError:
        print("Entrada no válida. Por favor, ingrese un número entero.")

#Ingreso de los niveles de oxígeno
for i in range(cantidad_mediciones):
    while True:
        try:
            oxigeno = float(input(f"Ingrese el nivel de oxígeno disuelto para la medición {i+1} (4.0 a 10.0 mg/L): "))
            if validar_oxigeno(oxigeno):
                niveles_oxigeno.append(oxigeno)
                break
            else:
                print("Nivel de oxígeno inválido. Por favor, ingrese un valor entre 4.0 y 10.0 mg/L.")
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número.")

#Cálculo del promedio de oxígeno
suma_oxigeno = 0
for oxigeno in niveles_oxigeno:
    suma_oxigeno += oxigeno

promedio_oxigeno = suma_oxigeno / len(niveles_oxigeno)

#Cálculo de las mediciones bajo 5.0 mg/L, entre 5.0 y 7.0 mg/L, y sobre 7.0 mg/L
mediciones_bajo_5 = 0
mediciones_entre_5_y_7 = 0
mediciones_sobre_7 = 0

for oxigeno in niveles_oxigeno:
    if oxigeno < 5.0:
        mediciones_bajo_5 += 1
    elif 5.0 <= oxigeno <= 7.0:
        mediciones_entre_5_y_7 += 1
    else:
        mediciones_sobre_7 += 1

#Mostrar resultados
print(f"\nPromedio de oxígeno disuelto: {promedio_oxigeno:.2f} mg/L")
print(f"Cantidad de mediciones bajo 5.0 mg/L (nivel crítico): {mediciones_bajo_5}")
print(f"Cantidad de mediciones entre 5.0 y 7.0 mg/L (nivel aceptable): {mediciones_entre_5_y_7}")
print(f"Cantidad de mediciones sobre 7.0 mg/L (nivel óptimo): {mediciones_sobre_7}")
```

### Ejercicio 6: Control de asistencia de trabajadores en faena minera

En una faena minera del norte de Chile, se necesita registrar la asistencia diaria de los trabajadores para efectos de control operativo.

Desarrolla un script en Python que permita gestionar esta información.

El programa debe:
- Registrar trabajadores usando un diccionario, donde:
    + clave = nombre del trabajador
    + valor = estado de asistencia ("presente" o "ausente")
- Permitir ingresar trabajadores de forma repetitiva hasta que el usuario escriba "fin".
- Validar que el estado ingresado sea solo "presente" o "ausente".

Al finalizar, el programa debe mostrar:
- Cantidad total de trabajadores registrados
- Cantidad de trabajadores presentes
- Cantidad de trabajadores ausentes
- Lista de trabajadores ausentes

Consideraciones:
- Usar un ciclo de duración desconocida (while) para el ingreso de datos.
- Usar un for para recorrer el diccionario y hacer los conteos.
- Separar claramente la etapa de registro y la etapa de análisis.


```python
'''

Control de asistencia en una faena minera.

El código debe generar un diccionario donde: 
    - clave = nombre del trabajador
    - valor = estado de asistencia ("presente" o "ausente")

El registro de asistencia finaliza cuando el usuario ingresa "fin" como nombre de trabajador.

Además, se debe validar que el estado de asistencia ingresado sea solo "presente" o "ausente".

Se debe mostrar:
    - Cantidad total de trabajadores registrados
    - Cantidad de trabajadores presentes
    - Cantidad de trabajadores ausentes
    - Lista de trabajadores ausentes.
    
Consideraciones: 
    -Se debe separar claramente la etapa de registro y la etapa de análisis.
'''

print("\n--- Control de Asistencia en Faena Minera ---")

#Función para validar el estado de asistencia
def validar_asistencia(asistencia):
    if asistencia.lower() in ["presente", "ausente"]:
        return True
    else:
        return False

#Diccionario para almacenar el registro de asistencia
asistencia = {}

#Registro de asistencia
while True:
    nombre_trabajador = input("Ingrese el nombre del trabajador (o 'fin' para terminar): ")
    if nombre_trabajador.lower() == "fin":
        break
    while True:
        estado_asistencia = input(f"Ingrese el estado de asistencia de {nombre_trabajador} ('presente' o 'ausente'): ")
        if validar_asistencia(estado_asistencia):
            asistencia[nombre_trabajador] = estado_asistencia.lower()
            break
        else:
            print("Estado de asistencia inválido. Por favor, ingrese 'presente' o 'ausente'.")

#Análisis de asistencia
total_trabajadores = len(asistencia)
trabajadores_presentes = 0
trabajadores_ausentes = 0
lista_ausentes = []

for nombre, estado in asistencia.items():
    if estado == "presente":
        trabajadores_presentes += 1
    else:
        trabajadores_ausentes += 1
        lista_ausentes.append(nombre)

#Mostrar resultados
print(f"\n--- Resultados ---")
print(f"Total de trabajadores registrados: {total_trabajadores}")
print(f"Trabajadores presentes: {trabajadores_presentes}")
print(f"Trabajadores ausentes: {trabajadores_ausentes}")
print(f"Lista de trabajadores ausentes: {', '.join(lista_ausentes)}")
```
