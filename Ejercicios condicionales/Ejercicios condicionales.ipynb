### 2.0 Escriba un algoritmo que pida y compare 2 números e identifique cuál es el mayor o si son iguales

```python
# Algoritmo que pida y compare 2 números e identifique cuál es el mayor o si son iguales

num1 = int(input("Ingrese un número 1"))

num2 = int(input("Ingresa un segundo número"))

if num1 > num2:
    print("El número", num1, "es el mayor.")
elif num2 > num1:
    print("El número", num2, "es el mayor.")
else:
    print("Los números son iguales.")

```

```python
# Mismo algoritmo, pero While para validar que los números no sean iguales.

num1 = int(input("Ingrese un número 1: "))
num2 = int(input("Ingresa un segundo número: "))

while True: 
    if num1 > num2:
        print("El número", num1, "es el mayor.")
        break
    elif num2 > num1:
        print("El número", num2, "es el mayor.")
        break
    else:
        print("Los números son iguales. Por favor, ingrese números diferentes.")
        num1 = int(input("Ingrese un número 1: "))
        num2 = int(input("Ingresa un segundo número: "))
```

### 2.1 Escriba un algoritmo que pida un número e identifique si es positivo, negativo o cero.

```python
# Algoritmo para ver si un número es positivo, negativo o cero.

num = int(input("Ingrese un número: "))

if num < 0: 
    print("El número es negativo.")
elif num > 0:
    print("El número es positivo.")
else:
    print("El número es cero.")
```

```python
# Mismo algoritmo, pero con validación de que el número ingresado sea realmente un número.

while True:
    try:
        num = int(input("Ingrese un número: "))
        break
    except ValueError:
        print("Entrada no válida. Por favor, ingrese un número entero.")

if num < 0:
    print("El número ", num, " es negativo.")
elif num > 0:
    print("El número ", num, " es positivo.")
else:
    print("El número ", num, " es cero.")
```

### 2.2 Escriba un algoritmo que pida 3 números al usuario, e identifique el mayor de ellos

```python
# Algoritmo para determinar el número mayor dentro de tres números.

print("Ingrese tres números para determinar cuál es el mayor.")

num1 = int(input("Ingrese el primer número: "))
num2 = int(input("Ingrese el segundo número: "))
num3 = int(input("Ingrese el tercer número: "))

if num1 > num2 and num1 > num3:
    print("El número ", num1, " es el mayor.")
elif num2 > num1 and num2 > num3:
    print("El número ", num2, " es el mayor.")
elif num3 > num1 and num3 > num2:
    print("El número ", num3, " es el mayor.")
else:
    print("Hay números iguales entre los ingresados.")
```

```python
# Algoritmo con la misma función, pero con validación de que los números ingresados sean realmente números, además, que se permitan todos los números reales.

print("Ingrese tres números para determinar cuál es el mayor.")

while True:
    try:
        num1 = float(input("Ingrese el primer número: "))
        num2 = float(input("Ingrese el segundo número: "))
        num3 = float(input("Ingrese el tercer número: "))
        break
    except ValueError:
        print("Entrada no válida. Por favor, ingrese números reales.")

if num1 > num2 and num1 > num3:
    print("El número ", num1, " es el mayor.")
elif num2 > num1 and num2 > num3:
    print("El número ", num2, " es el mayor.")
elif num3 > num1 and num3 > num2:
    print("El número ", num3, " es el mayor.")
else:
    print("Hay números iguales entre los ingresados.")
```

### 2.3 Escriba si una persona es menor de edad, mayor de edad o de la tercera edad. (18 y 65 años)

```python
# Algoritmo para determinar si una persona es mayor de edad, menor de edad o tercera edad.

print("Ingresa tu edad para saber tu rango de edad")
print("Menor a 18 años = Menor de edad")
print("De 18 a 64 años = Mayor de edad")
print("65 años o más = Tercera edad")

while True:
    while True:
        try:
            edad = int(input("Ingrese su edad: "))
            break
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número entero positivo.")

    if edad >= 0 and edad < 18:
        print("Usted es menor de edad.")
        break
    elif edad >= 18 and edad < 65:
        print("Usted es mayor de edad.")
        break
    elif edad >= 65:
        print("Usted está en la tercera edad.")
        break
    else:
        print("Edad no válida. Por favor, ingrese una edad positiva.")
```

### 2.4 Un tienda ofrece un descuento del 10% si la compra es mayor a $40.000, y del 20% si es superior a 100.000. Indique el valor a pagar.

```python
# Algoritmo para determinar el porcentaje de descuento de una tienda.

print("Ingrese el total de la compra para saber su descuento.")

while True:
    while True:
        try:
            total = int(input("Ingresa el total de la compra: "))
            break
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número entero positivo.")
            
    if total > 40000: 
        total_con_descuento = int(total * 0.9)
        print("Su total con descuento es: ", total_con_descuento)
        break
    elif total > 100000:
        total_con_descuento = int(total * 0.8)
        print("Su total con descuento es: ", total_con_descuento)
        break
    elif total >= 0 and total < 40000:
        print("No tiene descuento.")
        break
    else: 
        print("Total no válido. Por favor, ingrese un total positivo.")
```

### 2.5 Escriba un algoritmo que pida el peso en KILOS y la estatura de una persona en METROS, y entregue el índice de masa corporal (IMC).Además, muestre si la persona se encuentra bajo de peso (menor a $18.5$), normal ($18.5 -24.9$) o sobrepeso (25 ó más). $\text{IMC} = \text{peso} /(\text{estatura})²$

```python
# Algoritmo para calcular IMC (mejorado)
print("Ingrese su peso en kilogramos y su altura en metros.")

while True:
    entrada_valida = False
    while not entrada_valida:
        try:
            peso = float(input("Ingrese su peso: "))
            altura = float(input("Ingrese su altura: "))
            if peso <= 0 or altura <= 0:
                print("Peso y altura deben ser valores positivos mayores a cero.")
                continue
            entrada_valida = True
        except ValueError:
            print("Entrada no válida. Por favor, ingrese números reales para peso y altura.")
    
    IMC = peso / (altura ** 2)
    imc_redondeado = round(IMC, 2)
    
    if IMC < 18.5:
        print(f"Su IMC es: {imc_redondeado} y está en bajo peso.")
        break
    elif IMC <= 24.9:
        print(f"Su IMC es: {imc_redondeado} y está en peso normal.")
        break
    else:  # IMC >= 25
        print(f"Su IMC es: {imc_redondeado} y está en sobrepeso.")
        break
```
